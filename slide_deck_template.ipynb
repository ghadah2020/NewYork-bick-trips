```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
%matplotlib inline
import seaborn as sns
import seaborn as sb
import csv
import datetime as dt
import calendar
```

```python
# load in the dataset into a pandas dataframe
NewyYorkBike_clean = pd.read_csv('NewyYorkBike_clean.csv')
NewyYorkBike_clean.sample(5)
```

## #  Analyze New York Bike System display kind of user using the Bike Trips system in bar chart .
 After i see the bar chart the user from subscriber more than customer 


```python
#Display allmost kind of user using the Bike Trips system
# create the plot
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[0]
sb.countplot(data = NewyYorkBike_clean, x = 'usertype', color = base_color)

# add annotations
n_points = NewyYorkBike_clean.shape[0]
cat_counts = NewyYorkBike_clean['usertype'].value_counts()
locs, labels = plt.xticks() # get the current tick locations and labels

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = cat_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/n_points)
    plt.title(' Display kind of user using the Bike Trips system');
```

##   Analyze New York Bike System display the gender of user using the Bike Trips system in Pie chart .
after i see the Pie chart the user Male more than the female 

```python
# code for the pie chart seen above
plt.figure(figsize = [10, 5])
sorted_counts = NewyYorkBike_clean['gender'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 100,
        counterclock = False);
plt.axis('square');
plt.title('Display the gender of user they using  Bike Trips system');

```

# Analyze New York Bike System Display the higher months of users trips duration

when i see the chart i extract the most user they take long duration from age group (26-36 year) and the trip duration  increase in August and September.

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
NewyYorkBike_clean.groupby(['Month','Age_Grop']).count()['tripduration'].unstack().plot(ax=ax);
```

# Analyze New York Bike System Display the station is the most popular location for bike

when i see the  chart i extract  the most trips is start from station Grove st Path .

```python
plt.figure(figsize = [10, 10])
base_color = sb.color_palette()[0]
type_order=NewyYorkBike_clean ['start station name'].value_counts().index
sb.countplot(data = NewyYorkBike_clean, y = 'start station name', color = base_color,order=type_order);
plt.xticks(rotation=90);
```

# Analyze New York Bike System Display the tripsduration for Gender

```python
fig, ax = plt.subplots(figsize=(15,7))
ax = sb.barplot(data = NewyYorkBike_clean, x = 'usertype', y = 'tripduration', hue = 'gender')
ax.legend(loc = 7, ncol = 3, framealpha = 1, title = 'Gender')
plt.title('Display the tripsduration for Gender ');
```
